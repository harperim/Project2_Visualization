## 1. 분석배경
- 연도별 신고접수 건수 증감 추세 파악

### 아동학대 정의 및 성립요건
**아동학대란?**
- 보호자를 포함한 성인이 18세 미만의 미성년자 아이들의 건강이나 복지를 해하거나 제대로 된 성장발달을 못하도록 성적, 신체적 또는 정신적 폭력과 가혹행위가 있을 경우나 자식을 유기하거나 방임하는 경우 (아동복지법 제3조 제7호)

```python
# 신고접수 건수 데이터 병합

# 2001-2017 데이터
report1 = pd.read_csv('./data/신고접수_건수_2001_2017.csv',
                   encoding = 'cp949',  # 한글 처리 방식
                   low_memory = False)

# 2018-2021 데이터
report2 = pd.read_csv('./data/신고접수_건수_2018_2021.csv',
                   encoding = 'cp949',
                   low_memory = False)

# report1 열 이름 변경
report1.rename(columns = {report1.columns[0]: "접수유형별"}, inplace=True)
# report1

# report1 형식과 동일하게 report2 변경
new_report2 = report2.drop([2, 3])
new_report2 = report2.reindex([0, 1, 5, 4, 6])
new_report2.rename(index = {report2.index[5]: 2, report2.index[4]: 3, report2.index[6]: 4}, inplace=True)
# new_report2

# ----------------------------------------------------------------------------------------------------------

# 연도별 신고접수 건수 데이터 병합 
new_report2 = new_report2.drop(['접수유형별(1)', '접수유형별(2)'], axis=1)
report = report1.merge(new_report2, left_index=True, right_index=True)
report = report.T
report = report.drop(report.columns[[0, 2]], axis=1)
# report = report.drop([0, 2])
```

```python
# 컬럼 지정
report.columns=report.iloc[0]

# 첫행 제거
report.drop(labels='접수유형별', axis=0, inplace=True)
report.head()

# https://ehpub.co.kr/pandas-csv-dataframe-column-select-python/
```

```python
# 해외발생에서 -부분 -> 0으로 치환
report.replace('-', 0, inplace=True)

# 데이터타입 int로 변경
report['아동학대 의심사례'] = report['아동학대 의심사례'].astype('int')
report['동일신고'] = report['동일신고'].astype('int')
report['해외발생'] = report['해외발생'].astype('int')
report.info()

report.to_csv('./data/신고접수_건수_전체.csv')
```

```python
# 연도별 아동학대 의심사례 추세 그래프

suspicion = report['아동학대 의심사례']
suspicion
suspicion.plot(ls='--', lw=2, color='red', marker='.')

plt.title('2001-2021 연도별 아동학대 의심사례 신고접수 건수')
plt.xlabel('연도 (year)')
plt.ylabel('아동학대 관련 건수')
plt.ylim(0, 54000)
plt.legend()  # 범례

# 연도별 동일신고/해외발생 추세 그래프
sf_report = report[['동일신고', '해외발생']]
sf_report.plot(ls='--', lw=2, marker='.')

plt.title('2001-2021 연도별 동일신고/해외발생 건수')
plt.xlabel('연도 (year)')
plt.ylabel('동일신고/해외발생 관련 건수')
plt.legend()  # 범례
```

### <결론>
- 아동학대는 지속적으로 발생해 왔지만 의심사례 신고접수 건수가 13~14, 16년도 이후 급증했음을 알 수 있다. 당시 사회적 배경에 대해 알아보고자 한다.

#### 2013년
- **2013년 8월, 칠곡 아동학대 사건 (2013.08.14)**
    - 의붓어머니 임 씨는 2013년 8월 14일 오후 의붓딸 A양(사망 당시 만 8세)을 때린 뒤 복통을 호소하는 데도 병원에 데려가지 않아 장간막 파열에 따른 복막염으로 숨지게 한 혐의 등으로 구속 기소됐다. 또한 A양 언니(만 12세)에게 동생을 죽였다는 허위 진술을 강요하게 하여 공범으로 기소되게 하였으며, 추가 수사 과정에서 A양 언니는 공범이 아닌 피해자로 밝혀졌다. 임 씨는 A양 언니에게 말을 듣지 않는다며 세탁기에 가둬 돌리고, 성추행과 욕조에 가둬 물고문을 하였다. 또한 친아버지 김 모씨는 범행에 가담한 혐의로 함께 구속 기소되었다.
    - 출처 : 위키백과(https://ko.wikipedia.org/wiki/)
    
- **2013년 10월, 울산 아동학대 사건 (2013.10.24)**
    - 사건 당일은 이서현 양이 다니던 학교인 울산 호연초등학교에서 부산 아쿠아리움으로 소풍을 가기로 예정된 날이었다. 그다음 날 울산에서 인천으로 이사가 예정되어 있었고 이 양은 친구들과 마지막으로 꼭 이 소풍에 참여하고 싶어 했기에 원래 예정돼 있었던 이삿날까지 미뤄 둔 상태였다. 막상 당일 동거녀 박 모씨(당시 39세)는 이서현 양을 소풍에 보내주지 않았으며 이 양이 소풍을 보내 달라고 하자 사정없이 폭행하였다. 이로 인해 이 양은 갈비뼈 24개 중 무려 16개가 부러지는 치명상을 입고 부러진 갈비뼈에 폐를 다쳐 숨을 거뒀다. 
    - 출처 : 나무위키(https://namu.wiki/)

- **2014.09.29부터 아동학대범죄의 처벌 등에 관한 특례법 시행 (아동학대 특례법)**
    - 신고의무 확대 및 강화
    - 누구든지 아동학대범죄를 알게 되거나 의심하는 경우 아동보호전문기관이나 수사기관에 신고할 수 있다.
    - 신고의무자가 정당한 사유 없이 신고하지 않을 경우 부과되는 과태료를 300만원에서 500만원 이하로 상향
    - 출처 : 법률신문(https://www.lawtimes.co.kr/news/87079)
    - 참고 : 아동학대권리보장원 (http://www.korea1391.go.kr/new/page/rel_law.php)

#### 2016년
- **2015년 12월, 인천 11세 아동학대 사건 (2015.12.12)**
    - 인천광역시에서 2015년에 2년 동안 친부와 계모에 의해 감금되어 학대당하던 여아(당시 11세)가 탈출하여 발견된 사건이다. 이 아동이 장기 결석자로 밝혀지면서 전국 초등학교를 대상으로 장기 결석자 실태 조사가 이뤄졌고, 추가로 오랫동안 밝혀지지 않은 아동 학대 사건이 밝혀지게 되었다. 이 과정에서 부천 초등학생 토막 살인 사건, 부천 여중생 백골 살인 사건 등이 추가로 밝혀졌다.
    - 2015년 12월 12일에 인천에서 친부와 계모에 의하여 2년 동안 집에 감금되어 학대와 폭행을 당하던 11세 피해 여아 A는 너무 배가 고파 2층의 세탁실에서 가스 배관을 타고 집을 탈출하였다. 슈퍼마켓에서 과자를 훔쳐 먹던 A를 발견한 주인은 겨울에 반팔과 반바지를 입고 또래에 비해 지나치게 마른 모습을 수상하게 여겨 경찰에 신고했고, 피해자는 신고를 받고 출동한 경찰에 발견되어 구출되었다. 친부 B와 계모 C는 A가 탈출했다는 사실을 알고 달아났으나 경찰에 의해 검거되었다. 경찰은 B와 C에 대해 A를 감금하고 상습적으로 폭행한 혐의로 전격 구속하였고, 같은 집에 살면서 폭행에 가담하던 D도 같은 혐의로 전격 구속하였다.

#### 그래프 분석
- 아동학대 의심사례 신고접수는 감소한 연도 한 번 없이, 매년 꾸준히 증가하는 추세이다.   
    가장 큰 폭으로 상승한 해는 가장 최근 연도인 2021년으로, 매해 지남에 따라 아동학대 문제는 점차 심각해지고 있음을 알 수 있다.

- 해외에서 발생하는 아동학대는 국내에 비해 미미하게 발생하고 있다.


#### 결론
- 사회적 분위기와 법률 제정으로 인해 아동학대 사건 발생한 연도에 아동학대 의심사례 신고 접수는 급증했다. 하지만 곧바로 그다음 해에 증가폭이 감소한 것을 확인할 수 있고, 아동학대 사건이 발생한 연도에 기하급수적으로 접수 건수가 증가함을 알 수 있다. 어른들의 지속적인 관심과 국가적 차원에서의 지원이 필요하다고 생각된다.
