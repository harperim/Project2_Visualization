# 4. 아동학대 발생률 감소를 위한 노력

## 국가적 차원

### 1. 학대 피해 아동 지원을 위한 코디네이터 배치
성학대 피해아동이 사건 직후, 치료 또는 피해 보상을 받기 위해 직접 각각 기관을 찾아다니며 자신에게 필요한 지원 서비스를 신청해야 한다. 이 과정에서 지원 서비스에 대한 정보가 부족하거나 까다로운 지원규정으로 분명한 성학대 피해임에도 지원을 받지 못하는 경우가 생기면 피해아동과 부모는 그 상태로 지원받는 것을 포기하는 것이 현실이다.  
코디네이터는 성학대 피해아동의 수사과정부터 지원까지 피해아동을 전담해 도와줄 수 있는 사람으로 1차적으로 피해아동과 가족을 만나 피해사실을 명확하게 조사해 피해아동에게 수사 관련 정보나 지원정보를 제공하고, 피해아동과 가족을 면접해 필요한 지원기관으로 연계하도록 해야 한다. 전담 코디네이터를 통해 2차 피해를 방지하고 지원을 받지 못하는 피해아동이 없도록 도울 수 있을 것이다.  
이러한 지원은 성학대 피해아동에게만 국한되는 것이 아니라, 모든 학대 유형의 피해아동들에게 지원되어야 한다고 생각한다.

출처 : 베이비뉴스 (https://www.ibabynews.com/news/articleView.html?idxno=64356)

### 2. 가해자 나이/관계가 아닌 피해 수준에 따른 피해자 보호체계
현행 아동복지법은 가해자가 '성인'인 경우에만 미성년 친족성폭력 피해자는 아동 성학대 피해자로 보고, 보건복지부와 지방자치단체 아동학대 전담 공무원의 사례 관리와 회복지원을 받을 수 있다. 가해자도 '미성년자'라면 피해자는 아동 성학대 피해자로 인정받지 못하므로 친족성폭력 가해자가 성인인지 여부에 따라 피해자 보호체계가 다르게 작동한다.  
피해자는 피해 사실을 인정받지 못하고 또다시 가해자와 같은 공간에서 지내야 하는 것이 현시점에서 피해아동이 겪는 현실이므로 가해자의 나이에 따라 법을 적용시키는 것이 아닌, 피해 수준에 따라 가해자에게 처벌을 내려야 한다. 이로 인해 피해 사실을 인정받고, 피해아동과 학대행위자와의 분리를 우선시한 후 국가로부터의 지원을 받을 수 있도록 도움을 주어야 한다. 
또한 부모로부터의 신체/정서학대가 가장 많이 나타났는데, 아동의 보호자라 할지라도 피해 사실은 변함없기 때문에 피해아동과 학대행위자를 분리시키고 피해아동이 지낼 수 있는 환경을 마련해 주는 등 국가에서의 적극적인 인프라 구축이 필요하다.  
출처 : 한겨레신문 (https://www.hani.co.kr/arti/society/women/1049582.html)

## 개인적 차원

### 1. 피해아동 발견 시 바로 신고하기
아동학대에 해당되는 경우를 평소에 인지하고, 만약 유사한 상황을 목격하게 됐다면 바로 신고하도록 한다. 경찰, 아동보호전문기관, 아동보호담당고무원에게 신고할 수 있으며, 112, 117, 1391 등의 신고전화를 통해 신고할 수 있다.

### 2. 부모교육
아동학대의 예방을 목적으로 긍정적 훈육법에 대해 교육을 받도록 한다. 양육과정에서 발생할 수 있는 다양한 사례를 통해 아동학대 유형을 살펴보고, 통계적 수치를 통해 학대행위자, 학대발생장소, 학대행위자 특성 등을 확인하는 시간을 가져본다.  
또한 아이들을 체벌시키기보다 올바른 훈육태도를 통해 교육시키도록 하며, 지역 사회가 추구해야 할 아동 인권 보호 인식에 대해 학습해 보는 시간을 주기적으로 가져 경각심을 가지도록 한다.
