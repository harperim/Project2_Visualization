### 1. 시도별 신고접수 건수

```python
# 시도별 아동학대 신고접수 건수
province = pd.read_csv('./data/시도별_신고접수_건수_2018_2021.csv',
                      encoding = 'cp949',
                      low_memory = False)

# 총계 들어간 행 삭제
province.drop([0,1,2,3], axis=0, inplace=True)

province.drop(['계', '응급아동학대 의심사례'], axis=1, inplace=True)
province.head()
```

```python
p18 = province[province['시점']==2018]
p19 = province[province['시점']==2019]
p20 = province[province['시점']==2020]
p21 = province[province['시점']==2021]

p18 = p18.set_index('시도별(1)', drop=False)
p18.drop(['시도별(1)', '시점'], axis=1, inplace=True)
p18.rename(columns={'아동학대 의심사례': '2018'}, inplace=True)

p19 = p19.set_index('시도별(1)', drop=False)
p19.drop(['시도별(1)', '시점'], axis=1, inplace=True)
p19.rename(columns={'아동학대 의심사례': '2019'}, inplace=True)

p20 = p20.set_index('시도별(1)', drop=False)
p20.drop(['시도별(1)', '시점'], axis=1, inplace=True)
p20.rename(columns={'아동학대 의심사례': '2020'}, inplace=True)

p21 = p21.set_index('시도별(1)', drop=False)
p21.drop(['시도별(1)', '시점'], axis=1, inplace=True)
p21.rename(columns={'아동학대 의심사례': '2021'}, inplace=True)

p = pd.concat([p18, p19, p20, p21], axis=1)
p.head()

p.plot(kind='bar', figsize=(30, 10), width=0.8, alpha=0.8,
       color=['lightcoral', 'indianred', 'brown', 'maroon'])

plt.title('<시도별 아동학대 의심사례>', fontsize=30, fontweight='bold', pad=30)
plt.xlabel('시도별', fontsize=20, labelpad=15)
plt.ylabel('아동학대 의심사례 수 (건)', fontsize=20, labelpad=15)
plt.legend(loc='best', fontsize=20)
plt.xticks(rotation=0, fontsize=17)
plt.yticks(rotation=0, fontsize=17)
```

### 2. 아동학대 발생장소

```python
place = pd.read_csv('./data/아동학대_발생장소_2018_2021.csv',
                   encoding = 'cp949',
                   low_memory = False)

# 컬럼 이름 변경
place.rename(columns = {'가정내.1': '아동 가정 내', '가정내.2': '학대행위자 가정 내', 
                        '복지시설.1': '아동복지시설', '복지시설.2':'기타복지시설'}, inplace=True)

# 첫번째 줄 제거
place.drop(labels=0, axis=0, inplace=True)

# 발생건수가 높은 장소 5곳 추출 (평균값으로)
place = place.astype('int')
place['평균건수'] = place.mean(axis=1)  # 행별 장소에 따른 평균건수
place.loc[5] = place.mean(axis=0)  # 열별 장소에 따른 평균건수

place = place.astype('int')
place.rename(index={5: '평균 발생건수'}, inplace=True)
place.head()

# 상위 5개 장소
# sorted(place.loc['평균 발생건수'], reverse=True)
```

```python
# 행과 열에 위치한 평균건수 삭제
place = place.drop(['평균건수'], axis=1)
place = place.drop(['평균 발생건수'], axis=0)

# idx 설정
place = place.set_index('시점', drop=False)
place.drop(labels='시점', axis=1, inplace=True)

# 위에서 구한 상위 5개 장소만 df로 설정
place = place[['가정내', '아동 가정 내', '학교', '어린이집', '학대행위자 가정 내']]
# place

# plot 그리기
place.plot(ls='--', lw=2, marker='.')
plt.xticks([2018, 2019, 2020, 2021])
plt.title('연도별 아동학대 발생장소 추세', fontsize=15, fontweight='bold', pad=15)
plt.xlabel('연도 (year)', fontsize=10, labelpad=10)
plt.ylabel('아동학대 발생 건 수 (건)', fontsize=10, labelpad=10)
```

### 결론_2
- 각 행과 열에 위치한 평균건수를 제외한 상위 5개는 다음과 같다.
    - 가정내, 아동가정내, 학교, 어린이집, 학대행위자 가정 내
- 가정, 아동 가정 내에서 발생하는 건수는 연도에 따라 증가하고 있다. 학교, 어린이집, 학대행위자 가정 내에서 발생하는 건수는 비교적 줄어들고 있지만, 최근 들어 약간의 증가 추세를 보이고 있다.
- 아동학대 발생장소가 많이 일어나는 장소 상위 2곳은 가정내, 아동 가정 내로 모두 아동과 함께 사는 사람에 의해 학대가 일어남을 추측해 볼 수 있다.

### 3. 아동학대 유형별 현황

```python
# 아동학대 사례 유형별 현황
relation = pd.read_csv('./data/아동학대사례_유형별_현황_2018_2021.csv',
                      encoding = 'cp949',
                      low_memory = False)
relation = relation.T
relation = relation.reset_index()

relation.loc[1, 'index'] = '시점'
relation.loc[1, 0] = '학대유형'
relation.loc[2:7, 'index'] = '2018'
relation.loc[7:12, 'index'] = '2019'
relation.loc[12:17, 'index'] = '2020'
relation.loc[17:22, 'index'] = '2021'
relation.loc[1, 1] = '성별 소계'
relation.loc[1, 4] = '연령별 소계'
relation.loc[1, 12] = '관계 소계'

# 컬럼 변경
relation.columns = relation.iloc[1]
relation.drop(labels=[0, 1], axis=0, inplace=True)
```

```python
# 1. 성별
r_gender = relation[['시점', '학대유형', '성별 소계', '남아', '여아']]
r_gender = r_gender.set_index('시점', drop=False)
```

```python
# 성별, 연령별, 학대행위자 그래프로 나타내기

# 학대유형에 따른 피해아동 성별 비율

# 그래프 겹치지 않게 하기
plt.subplots(constrained_layout=True)

plt.subplot(221)
신체 = r_gender[r_gender['학대유형']=='신체학대']
신체 = 신체.astype({'성별 소계': 'int', '남아': 'int', '여아': 'int'})
신체.drop(['시점', '학대유형', '성별 소계'], axis=1, inplace=True)
신체.loc['총합', :] = 신체.loc['2018':'2021', :].sum(axis=0)
신체 = 신체.T
plt.title('<신체학대> 피해아동 성별 비율', fontsize=10, pad=2)
plt.pie(신체['총합'], autopct='%.2f%%', colors=["#dde5b6", "#adc178"], 
        labels=['남아', '여아'], labeldistance=0.3)

plt.subplot(222)
정서 = r_gender[r_gender['학대유형']=='정서학대']
정서 = 정서.astype({'성별 소계': 'int', '남아': 'int', '여아': 'int'})
정서.drop(['시점', '학대유형', '성별 소계'], axis=1, inplace=True)
정서.loc['총합', :] = 정서.loc['2018':'2021', :].sum(axis=0)
정서 = 정서.T
plt.title('<정서학대> 피해아동 성별 비율', fontsize=10, pad=2)
plt.pie(정서['총합'], autopct='%.2f%%', colors=["#dde5b6", "#adc178"], 
        labels=['남아', '여아'], labeldistance=0.3)

plt.subplot(223)
성 = r_gender[r_gender['학대유형']=='성학대']
성 = 성.astype({'성별 소계': 'int', '남아': 'int', '여아': 'int'})
성.drop(['시점', '학대유형', '성별 소계'], axis=1, inplace=True)
성.loc['총합', :] = 성.loc['2018':'2021', :].sum(axis=0)
성 = 성.T
plt.title('<성학대> 피해아동 성별 비율', fontsize=10, pad=2)
plt.pie(성['총합'], autopct='%.2f%%', colors=["#dde5b6", "#adc178"], 
        labels=['남아', '여아'], labeldistance=0.3)

plt.subplot(224)
방임 = r_gender[r_gender['학대유형']=='방임']
방임 = 방임.astype({'성별 소계': 'int', '남아': 'int', '여아': 'int'})
방임.drop(['시점', '학대유형', '성별 소계'], axis=1, inplace=True)
방임.loc['총합', :] = 방임.loc['2018':'2021', :].sum(axis=0)
방임 = 방임.T
plt.title('<방임> 피해아동 성별 비율', fontsize=10, pad=2)
plt.pie(방임['총합'], autopct='%.2f%%', colors=["#dde5b6", "#adc178"], 
        labels=['남아', '여아'], labeldistance=0.3)
```

### 1. 학대유형에 따른 성별 비율
- 학대유형은 총 4가지로 분류된다.
    - 신체학대, 정서학대, 성학대, 방임
- 신체학대, 정서학대, 성학대는 남아 여아의 비율에 큰 차이가 없는 것으로 보이지만, 성학대의 경우 피해아동이 여아인 경우 약 80%로 압도적인 수치임을 확인할 수 있다.

2. 형사처벌
- 형사처벌의 경우 보호처분에 비해 압도적으로 낮은 수치임을 확인할 수 있다. 성학대의 경우 보호처분보다 형사처벌의 판결이 더 많음을 확인할 수 있는데, 이마저도 최근 들어 감소하고 있는 추세이다.

아동학대 의심사례가 지속적으로 증가하고 있음에도 불구하고, 학대행위자들에 대한 처벌은 보호처분이 대부분이며, 형사처벌인 경우에도 벌금형에서 끝나는 일이 대다수이다. 이를 통해 아동학대 관련 처벌 수위가 높지 않음을 알 수 있다. 이는 사회문화적 특성이 많이 반영된 것으로 보이며, 가정 내에서 일어나는 범죄에 대한 처벌은 다른 범죄에 비해 비교적 더 많은 아량을 베푸는 것으로 보인다. 이에 대한 원인은 추후에 심층적으로 조사해 볼 필요가 있다고 생각된다.
